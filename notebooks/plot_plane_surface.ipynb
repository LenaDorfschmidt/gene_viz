### Demo notebook to add plane on mesh

The notebook shows how to use the function plot_surf_and_plane
It uses matplotlib_surface_plotting to plot a mesh and with an added plane on top

```python
# import libraries
import nibabel as nb
import numpy as np
import random

from gene_viz.viewer.plot_surf_and_plane import plot_surf_and_plane
```

```python
# load the brain mesh
gii_file= nb.load('../../michack_project_data/fs_LR.32k.L.pial.surf.gii')
brain_mesh={'coords':gii_file.darrays[0].data,
      'faces': gii_file.darrays[1].data}

# create a random overlay of the mesh
overlay_brain = np.array([random.choice(range(3)) for _ in range(len(brain_mesh['coords']))])

# load the MRI image
brain_img = nb.load('../../michack_project_data/MNI152_T1_1mm.nii.gz')

plot_surf_and_plane(mesh=brain_mesh, 
                    mesh_overlay=overlay_brain, 
                    mri_img=brain_img, 
                    slice_i=100,
                    slice_axes=1, 
                    rotate=[10, 80],
                    cmap = 'jet'
                    )
```
